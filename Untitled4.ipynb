```python
import pandas as pd
import random
import pandas as pd

df = pd.read_csv("kelimeler.csv", encoding="windows-1254", sep=';')
words_dict = df.set_index("English")["Turkish"].to_dict()

correct_counter = 0
wrong_counter= 0
word = ''
used_word_list = []

def new_word():
    global word

    if len(used_word_list) == len(words_dict):
        word_label.config(text="üéâ Game Over! üéâ")
        result_label.config(text=f"‚úÖ Doƒüru: {correct_counter} - ‚ùå Yanlƒ±≈ü: {wrong_counter}", fg="blue")

        answer.config(state='disabled')
        control_button.config(state='disabled')
        return

    while True:
        word = random.choice(list(words_dict.keys()))
        if word not in used_word_list:
            break

    used_word_list.append(word)
    word_label.config(text=f"What does '{word}' mean in Turkish?")

    result_label.config(text="")

    answer.delete(0,tk.END)

def check_it():
    global correct_counter, wrong_counter
    clean_answer = answer.get().strip().lower()

    if clean_answer == words_dict[word]:
        correct_counter +=1
        result_label.config(text="‚úÖ Doƒüru!", fg="green")
    else:
        wrong_counter +=1
        result_label.config(text=f"‚ùå Yanlƒ±≈ü! Doƒüru cevap: {words_dict[word]}", fg="red")

    score_label.config(text=f"Score: {correct_counter} Correct - {wrong_counter} Wrong")


    master.after(1500, new_word)


master = tk.Tk()

master.config(bg="#7db2d4")
master.title("English Turkish Word Game")
master.geometry("750x500")

word_label = tk.Label(master, text="", font=("Comic Sans MS", 25), bg="#7db2d4",fg="#140f26")
word_label.pack(pady=30)

answer = tk.Entry(master, font=("Verdana",20))
answer.pack(pady=25)

control_button = tk.Button(master, text="Check the Answer", font=("Verdana", 15), bg="#4CAF50", fg="white", command=check_it)
control_button.pack(pady=15)

result_label = tk.Label(master,text="", font=("Verdana", 12),bg="#7db2d4")
result_label.pack(pady=10)

score_label = tk.Label(master, text= "Score: 0 Correct- 0 Wrong", font=("Comic Sans MS",30), fg='#1a262e', bg="#7db2d4")
score_label.pack(pady=30)

new_word()

master.mainloop()
```
